```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
dftrain = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
dfeval = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/eval.csv')
y_train = dftrain.pop('survived')
y_eval = dfeval.pop('survived')
dftrain
```

```python
from sklearn.preprocessing import LabelEncoder, LabelBinarizer, MinMaxScaler
int_enc = LabelEncoder()
onehot_enc = LabelBinarizer()
df = dftrain.copy()
```

```python
ord_vars = ['class', 'deck']
bin_vars = ['sex', 'alone']
```

```python
def dummy_encoder(vars, enc, nominal=False):
    for var in vars:
        if nominal == True:
            df[df[var].unique()] = enc.fit_transform(df[var])
        else:
            df[var] = enc.fit_transform(df[var])
```

```python
dummy_encoder(ord_vars, int_enc)
dummy_encoder(bin_vars, int_enc)
df['class'] = df['class']+1
df[['age', 'fare']] = MinMaxScaler().fit_transform(df[['age', 'fare']])
df['age'].plot.hist()
df['family_size'] = df['parch'] + df['n_siblings_spouses']
df = df.drop(columns=['embark_town', 'parch', 'n_siblings_spouses'])
```

```python
corr_matrix = pd.concat([df, y_train], axis=1).corr()
sns.heatmap(corr_matrix, mask=np.tril(corr_matrix), annot=True)
```

```python
sex_group = pd.concat([df, y_train], axis=1).groupby('family_size')
sex_group['survived'].mean().plot(kind='bar')
```
